```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
file_path = "../data/raw/sample_-_superstore.xls" # ..→ para salir de la carpeta notebooks
```

```python
# Column names will be standardized (spaces, hyphens, and slashes removed)
df.columns = (
    df.columns
    .str.lower()
    .str.strip()
    .str.replace(" ", "_")
    .str.replace("/", "_")
    .str.replace("-", "_")
)

df.columns
```
